â€œGiven marks of students in multiple subjects, calculate:

subject-wise average

student-wise total

highest scorer

passing students

```python
import numpy as np
# Our Dataset
marks = np.array([
 #Math,Scie,Eng
    [80,70,60],   # student1
    [90,85,88],   # student2
    [55,60,65],   # student3
    [30,40,50]    # student4
])
names = np.array(["Ali","Bilal","Shakir","Ahmad"])
print(marks)

# subject-wise average 
print("Subject Wise average: ", marks.mean(axis = 0)) #Q1

# student-wise total
print("Student Wise Total: ", marks.sum(axis = 1))

# Highest score per subject
print("Highest score per subject: ",marks.max(axis = 0))

# Lowest score per subject
print("Lowest score per subject: ",marks.min(axis = 0))

# Overall topper
print("Overall topper: ",names[marks.sum(axis = 1).argmax()], " = ",marks.sum(axis = 1).max())

# passing students
passed_students = marks[np.all(marks>40,axis = 1)]
print("passing students: ",passed_students)

# failed students
failed_students = marks[np.any(marks<40,axis = 1)]
print("failed students: ",failed_students)
```
