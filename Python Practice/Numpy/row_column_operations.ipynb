# ⭐ Next: Concept – Axis (axis=0, axis=1)

Axis basically batata hai ki operation row pe apply ho, ya column pe.

axis = 0 ➜ down the rows

Combine vertically
(operate column-wise)

axis = 1 ➜ across the columns

Combine horizontally
(operate row-wise)

Example:mple:

```python
import numpy as np 
arr = np.array([[1,2,3],
                [4,5,6]])
print(arr)

print("Sum along column (axis = 0): ",arr.sum(axis = 0))

print("Sum along rows (axis = 1): ", arr.sum(axis = 1))
```

# ⭐ Next: try these (super important)

Given:

```python
arr = np.array([[1,2,3],
                [4,5,6]])

print("Mean:-->col wise ", arr.mean(axis = 0))
print("Min element col wise: ",arr.min(axis=0))
print("Max element col wise: ",arr.max(axis=0))

print("Mean:-->Row wise ",arr.mean(axis=1))
print("Min element row wise: ",arr.min(axis=1))
print("Max element row wise: ",arr.max(axis=1))
```
