## a) Write a function to normalize a list of numbers.
#### Normalization ka matlab numbers ko 0 aur 1 ke beech scale karna.

#### Formula (Min-Max Normalization):
### ð‘‹ð‘›ð‘œð‘Ÿð‘š = ð‘‹âˆ’ð‘‹ð‘šð‘–ð‘›/ð‘‹ð‘šð‘Žð‘¥âˆ’ð‘‹ð‘šð‘–ð‘›

```python
def normalize(data):
    min_val = min(data)
    max_val = max(data)
    normalized = [(x-min_val)/(max_val - min_val) for x in data] #list comprehension
    return normalized
nums = [10, 20, 30, 40]
print("Normalized:", normalize(nums))
```

## b) Create a function to calculate accuracy given TP, TN, FP, FN.
### TP --> True Positive -->True Prediction
### TN --> True Negative -->True Prediction
### FP --> False Positive -->False Prediction
### FN --> False Negative -->False Prediction

```python
def accuracy (TP,TN,FP,FN):
    total = TP+TN+FP+FN
    if total == 0:
        return 0
    else:
        return (TP+TN)/total
    
# let's call the accuracy function
print("Accuracy: ",accuracy(50, 45, 5, 5))
```

# c) Import statistics module and compute dataset statistics
### Helpful functions:

#### statistics.mean()-->Mean

#### statistics.median()-->Median

#### statistics.mode()-->Mode

#### statistics.stdev()-->Standard Deviation

#### statistics.variance()-->Variance

```python
import numpy as np 
import statistics as stats
dataset = np.random.randint(10,100,50)
dataset = dataset.tolist()

mean = stats.mean(dataset)

median = stats.median(dataset)

mode = stats.multimode(dataset)  #mode method for single mode, but i have kept random numbers, so there might exist more than one mode in a dataset

standard_deviation = stats.stdev(dataset)

variance = stats.variance(dataset)

print("Rnadom dataset: ", dataset)
print("Mean: ",mean)
print("Median: ",median)
print("Mode: ",mode)
print("Standard Deviation: ",standard_deviation)
print("Variance: ", variance)
```
