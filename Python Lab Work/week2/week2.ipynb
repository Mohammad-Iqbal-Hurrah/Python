## Program to calculate mean, median and mode of three numbers

```python
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
c = float(input("Enter third number: "))
nums = [a,b,c]
mean_val = sum(nums)/3
median_val = sorted(nums)[1] #3 gives middle value 
if a == b or a == c:
    mode_val = a
elif b == c:
    mode_val = b
else:
    mode_val = "No Mode"
print("Mean: ",mean_val)
print("Median: ",median_val)
print("Mode: ",mode_val)
```

## Convert temperature dataset from celsius to farenheit
## Formula F = (C*9/5)+32

```python
celsius_dataset = [0,25,37,100,-10]
farenheit_dataset = [(temp*9/5)+32 for temp in celsius_dataset] #list comprehension
print("Temperature in celsius: ", celsius_dataset)
print("Temperature in farenhiet: ",farenheit_dataset)
```

## Parse a string of comma separated numbers into a list of floats

```python
dataset = input("Enter comma separated numbers: ")
float_dataset = [float(x) for x in dataset.split(",")] #split() converts string ->list of substrings 
print("List of floates: ", float_dataset)
```
