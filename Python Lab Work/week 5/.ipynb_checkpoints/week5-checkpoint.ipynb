# a) Store student scores in a list and compute top 3 score

```python
scores = []
for i in range(0,5):
    scores.append(int(input("Enter marks: ")));
print(scores)
scores.sort(reverse = True)
print("Top three student marks: ",scores[:3])
```

# b) Represent (x,y) coordinates of points in a tuple and compute Euclidean distances. 

ðŸ“Œ Theory

A point in 2D space â†’ (x, y)
Euclidean distance between points A(x1,y1) and B(x2,y2):

distance = sqrt{(x2-x1)**2 + (y2-y1)**2}
	â€‹


```python
import math as m
A = (2,3)
B = (7,11)

x1,y1 = A
x2,y2 = B

dist = m.sqrt((x2 - x1)**2 + (y2 - y1)**2)

print("Point A: ",A)
print("Point B: ",B)
print("Euclidean distances: ",dist)
```
