# Mini Project using numpy and matplotlib

```python
import numpy as np

import matplotlib.pyplot as plt

temp_dataste = np.array([12,14,15, 13, 16, 18, 17]) #temperature dataset 
print(temp_dataste)

mean_temp = np.mean(temp_dataste)
max_temp = np.max(temp_dataste)
min_temp = np.min(temp_dataste)

print(mean_temp)
print(max_temp)
print(min_temp)

days = np.arange(1, 8)

print(days)

#title of the graph
plt.title("1 Week Temperature Analysis")

#daily temperature of a week
plt.plot(days,temp_dataste,marker='o', label="Daily Temperature")

#mean temperatur 
plt.axhline(mean_temp, color='black', linestyle='--',label=f"Mean Temp = {mean_temp:.2f}C")

#highest temperature
plt.scatter(days[np.argmax(temp_dataste)], max_temp, color='red', s=200,label="Highest Temp")

#Normal temperatur 
plt.fill_between(days, mean_temp - 1, mean_temp + 1,color='yellow', alpha=0.4, label="Comfort Zone")

plt.fill_between(days, mean_temp-2,mean_temp + 2,color='red', alpha=0.3, label="Unhealthy Zone")


#x-axis label
plt.xlabel("Day")
# y-axis label
plt.ylabel("Temperature (C)")

plt.grid(True)
plt.legend()
```
