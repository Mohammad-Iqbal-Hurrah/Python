# program to count how many values in a dataset are 

```python
above/below average. 
import numpy as np
dataset = np.random.randint(10,100,50)

average = sum(dataset)/len(dataset)
above = 0
below = 0
equal = 0
for data in dataset:
    if data>average:
        above +=1
    elif data<average:
        below +=1
    elif data == average:
        equal +=1
print("Data set: ",dataset)
print("Average: ", average)
print("Values above average are: ",above)
print("Values below average are: ",below)
print("Values that are equal to average are: ",equal)
```

## first 10 terms of the Fibonacci sequence and store in a list. 
## 0 1 1 2 3 5 8 13 21 32.... --> fibonacci sequence--> next term = sum of previous consective two terms 

```python
fibonacci_sequence = [0,1] #first two terms we already know 
for i in range(2,10):
    next_term = fibonacci_sequence[i-1] + fibonacci_sequence[i-2]
    fibonacci_sequence.append(next_term)
print("First 10 terms of a fibonacci sequence are: ", fibonacci_sequence)
```

## simple frequency histogram (using “*”) for given 
## categorical data.

```python
dataset = ['A','B','A','C','B','A']
print(dataset)
freq = {} #Empety Dictionary
for item in dataset:
    if item in freq:
        freq[item] += 1 #if the item already exists then increase its frequency by 1
    else:
        freq[item] = 1 #if the item item is commin gfor the first time, then its frequency is 1

for key, value in freq.items():  
    print(key + " : " + "*" *value)
print("Frequency of A: ",freq['A'])
print("Frequency of B: ",freq['B'])
print("Frequency of C: ",freq['C'])
print(freq)
print(freq.items()) #returns a list of tuples including key value pairs 
```
