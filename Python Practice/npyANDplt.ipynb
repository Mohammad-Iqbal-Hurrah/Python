```python
import numpy as np 
import matplotlib.pyplot as plt 
dataset = np.random.randint(10,100,50) #Generate 50 random numbers from 10 to 100
print(dataset)
mean_val = np.mean(dataset)
max_val= np.max(dataset)
min_val = np.min(dataset)
print("Mean = ", mean_val)
print("Min = ", min_val)
print("Max = ", max_val)
plt.plot(dataset, label = 'Random data')
plt.axhline(mean_val, color = 'red',linestyle = '--', label = f"Mean = {mean_val: .2f}") #Horizontal line draw 
plt.title("Random Data plot using numpy and matplotlib practice...") #Title of the plot 
plt.xlabel("index") #x-axis label
plt.ylabel("value")
plt.legend() #Labes are displayed using this legend function
plt.grid(True) #grid type background 
plt.show()
```
