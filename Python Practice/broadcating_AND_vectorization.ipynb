# ‚≠ê Broadcasting & Vectorized Operations
### üí° Broadcasting
Broadcasting matlab:

Jab NumPy different shape ke arrays ko ek sath calculation karwata hai automatically adjust karke

Broadcasting tab chalta hai jab shapes match ho jaye ya automatically expand ho sake

#### Broadcasting tab hoti hai jab:
same OR 1


‚úî 1. Size is same

OR

‚úî 2. One of them is 1


#### No broadcasting jab:
different AND both > 1

```python
import numpy as np
arr = np.array([1,2,3,4,5])
print("Array without broadcasting: ",arr)
arr = arr+5 #Broadcasting
print("Array after broadcasting: ",arr) #-->[1+5,2+5,3+5,4+5,5+5]-->[6,7,8,9,10]
```

### ‚≠ê Vectorized operations
Means:

operations element-by-element perform hote hain bina loops likhe

```python
print("Vectorized array: ",arr*5)
```

# Incorrect Broadcasting 
shape of a -->(1,3)
shape of b --> (1,2)-->it should be either 1,1 or 1,3

```python
a = np.array([[1,2,3]])
b = np.array([[10,20]])

print(a.shape)
print(b.shape)

print(a + b) #it will throq an error
```
