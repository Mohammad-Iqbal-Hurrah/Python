# ⭐ Concept 6 — Fancy Indexing

Fancy indexing =
Indexing using another list/array of indexes.

Basically:

we choose multiple indices simultaneously
### ⭐ Example 1 — Pick specific elements

```python
import numpy as np 

arr = np.array([10,20,30,40,50])
print("Array: ", arr)

print("Accessing multiple indices simultaneouslt: ",arr[[0,2,4]]) #we are accessing elements which are indices at 0,2 and 4-->values = 10,30 and 50
```

### ⭐ Example 2 — Out of order
means that the order is your own hands

you decide the order 

```python
print(arr[[4,0,2]])
```

# ⭐ Fancy indexing in 2D

```python
arr = np.array([[10,20,30],
                [40,50,60],
                [70,80,90]])
print("2D Array is:", arr)

fancy_array = arr[[0,2],[1,2]]

#(x1,y1),(x2,y2)
print("2D Array fancy indexing: ",fancy_array) #select element at index 0,1-->20 and 2,2--> 90
```

# ⭐ Concept: Row & Column separately

```python
#ROWS 
print("Row 0 and 2: ", arr[[0,2], :]) #row number 0 and 2
print("Row 2 and 0: ",arr[[1,0],:]) #row number 2 and 0

#ROWS 
print("column 0 and 1: ",arr[:,[0,2]]) #column number 0 and 2
print("column 1 and 2: ",arr[:,[1,2]]) #column number 0 and 2
```

# ⭐ Mixed slicing + fancy indexing

```python
print("first row first element and 3 element and 3rd  row first element and 3rd element: ",arr[[0,2], 1:3]) #first row first element and 3 element and 3rd  row first element and 3rd element
```

# ⭐ Fancy is DIFFERENT from slicing

#### Slicing

##### continuous part
##### must be ordered
##### uses :

#### Fancy
##### selected elements
##### can be random
##### uses lists

# ⭐ Mini Practice — your turn
##### GIVEN: [[11,22,33,44], [55,66,77,88]  [99,100,111,1 ]]

##### Find using fancy indexing:

##### 1️⃣ elements at positions (0,0), (1,2), (2,3)
##### 2️⃣ pick rows 0 and 2
##### 3️⃣ pick columns 1 and 3
##### 4️⃣ pick [11, 77, 122] using only fancy indexing
##### 5️⃣ pick 4th column using fancy indexing (not slicing)]]

```python
arr = np.array([[11,22,33,44],[55,66,77,88],[99,100,111,122]])
print("Array is: ",arr)

# 1️⃣ elements at positions (0,0), (1,2), (2,3) --> [11,77,122]
print("Question 1: ",arr[[0,1,2],[0,2,3]])


# 2️⃣ pick rows 0 and 2
print("Question 2 Row 0 and 2: ",arr[[0,2],:])

# 3️⃣ pick columns 1 and 3
print("Question 3 Column 1 and 3: ",arr[:,[1,3]])

# 4️⃣ pick [11, 77, 122] using only fancy indexing
print("Question 4: ",arr[[0,1,2],[0,2,3]])


# 5️⃣ pick 4th column using fancy indexing (not slicing)
print("Question 5: ",arr[[0,1,2],[3,3,3]])
```
